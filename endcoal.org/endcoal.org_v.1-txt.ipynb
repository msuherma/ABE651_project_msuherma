```python
boxprops = dict(linestyle='-', linewidth=4, color='k')
medianprops = dict(linestyle='-', linewidth=4, color='r')
ax = df.boxplot(column=['Indonesia',
             'Vietnam',
             'Malaysia',
             'Philippines',
             'Laos',
             'Cambodia',
             'Brunei',
             'Myanmar',
             'Thailand'],
                showfliers=False, showmeans=True,
                boxprops=boxprops,
                medianprops=medianprops,
                rot=-45)
plt.xlabel("Country", size =14)
plt.ylabel("Capacity, MW", size=14)
plt.title(f'ASEAN New Coal-fueled Power Plants in the 2008-2019', size=18)
plt.savefig("ASEAN New Coal-fueled Power Plants in the 2008-2019.png")
```

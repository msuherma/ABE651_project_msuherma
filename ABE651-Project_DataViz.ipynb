```python
def country_boxplot(country_abbr, country, x_ax, y_ax):
    """ country_abbr = country abbreviation
        county = country for the title of plot and name of the PNG file
        x_ax = x-axis
        y_ax = y-axis
    """
    plt.style.use('seaborn-talk')
    fig, ax = plt.subplots(figsize=(10,10))
    ax.set(yscale="log")
    sns.boxplot(y=y_ax, x=x_ax, data=country_abbr, width=0.3, palette="Accent")
    sns.stripplot(y=y_ax, x=x_ax,data=country_abbr, jitter=True,dodge=True, marker='o', alpha=0.5,color='grey')
    plt.xlabel(x_ax.capitalize(), size =14)
    plt.ylabel(y_ax.capitalize() + " (log-scale)", size=14)
    plt.title(f'{y_ax.title()} (log-scale) vs  {x_ax.title()} of {country}', size=18)
    plt.savefig(country + "_Plant_Capacity_vs_Fuel" ".png")
    return ;
```

```python
def country_boxplot2(country_abbr, country, x_ax, y_ax):
    """ country_abbr = country abbreviation
        county = country for the title of plot and name of the PNG file
        x_ax = x-axis
        y_ax = y-axis
    """
    plt.style.use('seaborn-talk')
    fig, ax = plt.subplots(figsize=(10,10))
    sns.boxplot(y=y_ax, x=x_ax, data=country_abbr, width=0.3, palette="Accent")
    sns.stripplot(y=y_ax, x=x_ax,data=country_abbr, jitter=True,dodge=True, marker='o', alpha=0.5,color='grey')
    plt.xlabel(x_ax.title(), size =14)
    plt.ylabel(y_ax.title() + " ", size=14)
    plt.title(f'{y_ax.title()} vs  {x_ax.title()} of {country}', size=18)
    plt.savefig(country + "_Commissioning_year_vs_Fuel" ".png")
    return ;
```

```python
country_boxplot2(idn, "Indonesia",'primary_fuel', 'commissioning_year')
```
